```python
import numpy as np
import mglearn
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import font_manager, rc
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from matplotlib import font_manager, rc
import chardet

font_name = font_manager.FontProperties(\
fname = "c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family = font_name)

with open('food_store1.csv', 'rb') as f:
     result = chardet.detect(f.read())
        
##name = np.loadtxt("C:/jupyter_project/food_store1.csv", delimiter = ",", dtype = "U")
##name = name[:,:]
data = pd.read_csv("C:/jupyter_project/food_store1.csv", sep = ",", encoding=result['encoding'])
```

```python
data.columns
```

```python
df1 = data[['한식', '중식', '일식', '양식', '아시안', '패스트푸드', '가격대(만원)']]
df1
```

```python
pca = PCA(n_components = 2)
x_2D = pca.fit_transform(df1)
```

```python
pd.DataFrame(x_2D)
```

```python
plt.scatter(x_2D[: , 0:1], x_2D[: , 1:])
```

```python
model = KMeans(n_clusters = 5)
model.fit(x_2D)
```

```python
model.labels_
data['Label'] = list(model.labels_)
data.to_csv("C:/jupyter_project/food_store1.csv", index = False, encoding='ms949')
```

```python
# 그래프 사이즈 설정
plt.figure(figsize=(14,7))
# 색상 배열
colormap = np.array(['red','lime', 'black', 'blue', 'gray'])
# 그래프 출력 (클러스터링)
plt.scatter(x_2D[: , 0:1], x_2D[: , 1:], c=colormap[model.labels_], s=40)
plt.title('클러스터링 결과')
```

```python
df2 = pd.read_csv("C:/jupyter_project/food_store1.csv", sep = ",", encoding='ms949')
y = df2[['Label']]
```

```python
# knn 알고르짐
# 새로운 데이터 포인트에 가장 가까운 k개의 이웃을 찾는다.
# 1개의 이웃만을 찾는다.
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(x_2D, y)
```

```python
## 사용자가 특정매장을 선택
x_test1 = np.array([[0.60476752, -0.4800025]])
x_test2 = np.array([[0.14042799,  0.2459149]])
x_test3 = np.array([[0.394313, 0.915605]])
```

```python
predList = []
predList.append(knn.predict(x_test1))
predList.append(knn.predict(x_test2))
predList.append(knn.predict(x_test3))

for index, value in enumerate(predList) :
    print("예측결과%d : %d" % (index+1, value))    
```

```python
selStore = []
for index, value in enumerate(predList)  :
    df3 = df2[df2.Label == int(value)]
    selStore.append(pd.DataFrame({"Label" : list(df3['Label']),
                                  "위치" : list(df3['위치']),
                                  "거리(m)" : list(df3['거리(m)']),
                                  "게이트거리(m)" : list(df3['게이트거리(m)']),
                                  "총거리(m)" : list(df3['총거리(m)']),
                                  "매장명" : list(df3['매장명'])}))
```

```python
##현재 위치에서 첫번째 경유지까지 거리순으로 정렬
a = selStore[0].sort_values(by='총거리(m)').head(5)
a
```

```python
name1 = list(a['매장명'])
name1
```

```python
## selStore[0]의 상점들과 selStore[1]의 각상점들 간의 거리를 계산해야한다.
## 거리순으로 정렬
cal1 = []
name2 = []
for index,i in enumerate(name1) :
    ## 아래의 path에 경유지간의 거리를 계산한 파일이 있어야 한다.
    path = "C:/jupyter_project/"+i+"_distance.csv"
    cal1.append(pd.read_csv(path, sep = ",", encoding='ms949').sort_values(by='총거리(m)').head(1))
    name2.append("".join(cal1[index]['매장명'].values))
```
