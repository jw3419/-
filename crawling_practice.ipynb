```python
from selenium import webdriver as wd
from bs4 import BeautifulSoup
import math
import re
import urllib.request as req
import requests
import http.cookiejar
import sys
import numpy as np
```

```python
site_url = "https://www.google.com/search?hl=ko&source=hp&ei=bzvGXJbjDIXyBAWX65bICQ&q={0}&start={1}"
```

```python
driver = wd.Chrome(executable_path="C:/chromedriver_win32/chromedriver.exe")
name = np.loadtxt("C:/jupyter_project/food.csv", delimiter = ",", dtype = "U")
name = name[:,0].tolist()
f = open("C:/jupyter_project/crawling_data/test.txt", 'w', encoding="utf-8")
```

```python
for list in name :
    list = "인천공항" + list
    print(list)
    print()
    try :
        for i in range(0, 2) :
            cur_url = site_url.format(list,(i*10))
            print("=========================================================================================================")
            print(list+str (i+1)+"page")
            driver.get(cur_url)
            driver.implicitly_wait(2)
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            sites = soup.select('div.TbwUpd')

            print()
            for link in sites : 
                fake = link.text.strip()
                if "https://" not in fake :
                    fake = "http://" + fake
            
                if "›" in fake : 
                    delete = fake.index('›')
                    fake = fake[:delete]
                print(fake)
            
                url = fake
                driver.get(url)
                driver.implicitly_wait(2)
                html = driver.page_source
                soup = BeautifulSoup(html, 'html.parser')

                if "naver" in fake :
                    info = soup.select('p.se_textarea')
                elif "tistory" in fake : 
                    info = soup.select('div.area_view > p > span > b > span')

                for i in info :
                    f.write(i.text.strip() + "\n")

        
            print()

  

    except Exception as e :
        print("오류발생", e)
        
f.close()
```
